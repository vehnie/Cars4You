```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_test= pd.read_csv('../data/test.csv')
```

```python
# distribution of the year
plt.figure(figsize=(10, 6))
sns.histplot(df_test['year'], bins=20, kde=True)
plt.title("Distribution of Year")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

```python
# scatter plot of year vs. mileage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='mileage', data=df_test, alpha=0.5)
plt.title("Year vs. Mileage")
plt.xlabel("Year")
plt.ylabel("Mileage")
plt.show()
```
